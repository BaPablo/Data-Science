```
from sklearn.datasets import load_breast_cancer
```

```
# Cargar el dataset

data = load_breast_cancer()
```

```
# Tipo de dato caragdo
print(type(data))
```

```
# Descripción del dataset cargado
print(data.DESCR)
```

```
# Dimensiones del dataset

filas, columnas = data.data.shape

print('El dataset contiene: {} filas y {} columnas'.format(filas,columnas))
```

```
# Atributos o features del dataset

print(data.feature_names)
```

```
data.target
```

```
# Targets o labels

print(data.target_names)
'''
0: Malignant
1: Benign
'''
```

```
import pandas as pd

print('Versión de pandas: ', pandas.__version__)
```

```
df_X = pd.DataFrame(data = data.data,columns=data.feature_names)
df_y = pd.DataFrame(data = data.target, columns=['target'])
print(df_X)
print(df_y)
```

```
dataset_df = df_X.join(df_y,how='right')
dataset_df
```

```
print(dataset_df.target.unique())
```
